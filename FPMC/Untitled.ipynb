```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def runningMeanFast(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]
```

```python
curve = np.loadtxt('fullModel-9June.csv',delimiter=',')
plt.figure(figsize=[15,7])
c1 = runningMeanFast(curve[:,1],1000)
plt.plot(c1)
plt.ylim([0.40, 0.48])
```
